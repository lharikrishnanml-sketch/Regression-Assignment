```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age','bmi','children','sex_male','smoker_yes']]
dep=dataset[['charges']]
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
# ULTRA-FAST: Only 12 combinations
param_grid = {
    'kernel': ['rbf'],                    # Only RBF (usually best for this data)
    'C': [100, 1000, 3000],              # 3 values
    'gamma': ['scale', 0.01],            # 2 values  
    'epsilon': [0.1, 1]                  # 2 values
}

# This gives: 1 × 3 × 2 × 2 = 12 combinations × 5 folds = 60 fits
# Should complete in under 1 minute!

grid = GridSearchCV(
    SVR(),
    param_grid,
    refit=True,
    verbose=3,
    n_jobs=-1,
    cv=5
)

grid.fit(indep, dep.values.ravel())
```

```python
grid
```

```python
re=grid.cv_results_
```

```python
#grid_predictions=grid.predict(X_test)
```

```python
#from sklearn.metrics import r2_score
#r_score=r2_score(y_test,grid_predictions)
print("The R_ score value for best parameter {}:".format(grid.best_params_))
```
